Indexing & slicing

subseting
- create a new table (subset) from our matrice
- transformate the subset

Boolean Indexing

- create a mask

- A[condition] = result

- filter the data

```python
import numpy as np
```

```python
A = np.random.rand(3,4)
A
```

```python
A.shape
```

```python
#first row : indexing
A[0]
```

```python
#first column : slicing
A[:, 0]
```

```python
#subset
sub = A[1:, 1:3]
sub
```

```python
#transform subset
A[1:, 1:3] = 1
A
```

```python
B = np.zeros((4,4))
B
```

```python
B[1:3, 1:3]
```

```python
B[1:3, 1:3] = 1
B
```

```python
C = np.ones((5,5))
C
```

```python
#slicing with one step
C[::2, ::2] = 55
C
```

```python
#from 1 to 100 with shape 10*10 
D = np.random.randint(1,100, [10,10])
```

```python
D
```

```python
#creat a mask
D<50
```

```python
#replace by 0
D[D<50] = 0
```

```python
D
```

```python
#filter numbers smal than 60
D[D<60]
```

```python
from scipy import misc
import matplotlib.pyplot as plt
```

```python
# creat object 'bob' from face 
bob = misc.face(gray=True)
```

```python
#3 dimention 
print(bob.size, bob.shape)
```

```python
#print bob
plt.imshow(bob, cmap=plt.cm.gray)
plt.show
```

```python
#zoome
h= bob.shape[0] #lignes
w = bob.shape[1] #colonnes

zoom_bob = bob[h//4 : -h//4, w//4 : -w//4]
plt.imshow(zoom_bob, cmap = plt.cm.gray)
plt.show
```

```python
zoom_bob[zoom_bob > 150] = 255
plt.imshow(zoom_bob, cmap = plt.cm.gray)
plt.show
```

```python
bob= bob[::2, ::2]
plt.imshow(bob, cmap=plt.cm.gray)
plt.show
```
