**Image Processing with Python **
- Image Processing using Python is a fun way to understand how pictures can be represented via math and code. I hope that this short article can give you an idea of how your machines understand image data.

```python
# Importez d'abord les biblioth√®ques Python requises 
# First import the required Python Libraries
import numpy as np
import matplotlib.pyplot as plt
from skimage import img_as_uint
from skimage.io import imshow, imread
from skimage.color import rgb2hsv
from skimage.color import rgb2gray
```

```python
array_1 = np.array([[255, 0], 
                    [0, 255]])
imshow(array_1, cmap = 'Pastel1_r')
```

```python
array_2 = np.array([[255, 0, 255], 
                    [0, 255, 0],
                    [255, 0, 255]])
imshow(array_2, cmap = 'gray');
```

```python
array_spectrum = np.array([np.arange(0,255,17), 
                   np.arange(255,0,-17),
                   np.arange(0,255,17),
                   np.arange(255,0,-17)])
fig, ax = plt.subplots(1, 2, figsize=(12,4))
ax[0].imshow(array_spectrum, cmap = 'gray')
ax[0].set_title('Arange Generation')
ax[1].imshow(array_spectrum.T, cmap = 'gray')
ax[1].set_title('Transpose Generation');
```

```python
array_colors = np.array([[[255, 0, 0], 
                         [0, 255, 0],
                         [0, 0, 255]]])
imshow(array_colors);
```

```python
hich = imread("/content/IMG_20220419_200021.jpg")
imshow(hich);
```

```python
hich.shape
```

```python
fig, ax = plt.subplots(1, 3, #one row 3 columns
                       figsize=(6,4), 
                       sharey= True)
    
ax[0].imshow(hich[:, 0:257])
ax[0].set_title('First Split')
    
ax[1].imshow(hich[:, 257:514])
ax[1].set_title('Second Split')
    
ax[2].imshow(hich[:, 414:771])
ax[2].set_title('Third Split');
```

```python
imshow(hich[100:700, 200:600])
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12,4), sharey = True)
ax[0].imshow(hich[:,:,0], cmap='Reds')
ax[0].set_title('Red')
ax[1].imshow(hich[:,:,1], cmap='Greens')
ax[1].set_title('Green')
ax[2].imshow(hich[:,:,2], cmap='Blues')
ax[2].set_title('Blue');
```

```python
# Additionally, we can convert the image from RGB (Red, Green, Blue) to HSV (Hue, Saturation, Value).
hich_hsv = rgb2hsv(hich)
fig, ax = plt.subplots(1, 3, figsize=(12,4), sharey = True)
ax[0].imshow(hich_hsv[:,:,0], cmap='hsv')
ax[0].set_title('Hue')
ax[1].imshow(hich_hsv[:,:,1], cmap='gray')
ax[1].set_title('Saturation')
ax[2].imshow(hich_hsv[:,:,2], cmap='gray')
ax[2].set_title('Value');
```

```python
hich_gray = rgb2gray(hich) #transform into 2 dimensions
uint = [0.75, 0.50, 0.25, np.mean(hich_gray)]

fig, ax = plt.subplots(1, 4, figsize=(17,6), sharey = True)
for i,j in enumerate(uint):
  ax[i].imshow(img_as_uint(hich_gray > j), cmap = 'gray')
  
```
